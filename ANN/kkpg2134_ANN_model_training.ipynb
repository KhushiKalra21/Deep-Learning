```python
# Download the ZIP file from Google Drive
import gdown

file_id = "1X40NeGmYe0epMXrewSVlbQscLaX4u9qT"
destination = "/content/dataset.zip"  # Save as dataset.zip

gdown.download(f"https://drive.google.com/uc?id={file_id}", destination, quiet=False)

# Extract the ZIP file
import zipfile
import os

extract_path = "/content/extracted_data"
os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(destination, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(" File downloaded and extracted successfully!")
```

```python
import pandas as pd
import os

# List all files in extracted folder
data_folder = "/content/extracted_data"
files = os.listdir(data_folder)
print(" Extracted Files:", files)

#Load the first CSV file (modify if needed)
csv_file = [f for f in files if f.endswith('.csv')][0]  # Picks the first CSV file
file_path = os.path.join(data_folder, csv_file)

#Read the CSV file
df = pd.read_csv(file_path)

#Display basic information
print(" Dataset Preview:")
display(df.head())  # Show first 5 rows

print("\n Dataset Info:")
df.info()  # Show column details

print("\n Missing Values:")
print(df.isnull().sum())  # Check for missing values
```

```python
import pandas as pd

# Assuming your original DataFrame is named 'df'
important_features = [
    'age', 'gender', 'income_bracket', 'loyalty_program', 'membership_years',
    'marital_status', 'number_of_children', 'education_level', 'occupation',
    'avg_purchase_value', 'purchase_frequency', 'last_purchase_date',
    'avg_discount_used', 'preferred_store', 'online_purchases',
    'in_store_purchases', 'avg_items_per_transaction', 'avg_transaction_value',
    'churned'  # Include the target variable
]

# Create a subset DataFrame
subset_df = df[important_features]

# Now 'subset_df' contains only the selected important features
```

```python
# Display the first few rows of the subset DataFrame
subset_df.head()
```

```python
from google.colab import files

subset_df.to_csv('subset_data.csv', index=False)
files.download('subset_data.csv')
```

## Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = subset_df.select_dtypes(include=['object']).columns

# Apply Label Encoding
for col in categorical_cols:
    le = LabelEncoder()
    subset_df[col] = le.fit_transform(subset_df[col])

print(" Categorical variables encoded.")
```

## Scale Numerical Features


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_cols = subset_df.select_dtypes(include=['int64', 'float64']).columns

subset_df[numerical_cols] = scaler.fit_transform(subset_df[numerical_cols])

print(" Numerical features scaled.")
```

## Define Features & Target Variable

```python
# Define target variable (modify based on your dataset)
target_col = "churned"  # Example target column, change as needed

X = subset_df.drop(columns=[target_col])  # Features
y = subset_df[target_col]  # Target variable

print(" Features and target variable defined.")
```

## Split into Train & Test Sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=552134)

print(" Data split into training and testing sets.")
```

## Import Required Libraries

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

## Build the ANN Model

```python
# Build ANN Model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile Model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train Model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

## Evaluate Model Performance

```python
# Evaluate Model
y_pred = (model.predict(X_test) > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
