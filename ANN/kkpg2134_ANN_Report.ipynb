## **3. Project Implementation**  

### **3.1 Dataset Used**  

The dataset used in this project was downloaded from **Kaggle**:  
[Large Retail Data Set for EDA](https://www.kaggle.com/datasets/utkalk/large-retail-data-set-for-eda)  

#### **Dataset Overview**  
- Contains **500,000+** customer records.  
- Includes **features such as customer demographics, transaction history, engagement levels, and churn status**.  
- The target variable is **"churned" (1 = churned, 0 = retained)**.  



### **3.2 Data Preprocessing & Cleaning**  

> **Steps Followed:**  

1. **Download & Extract Data**:  
   - The dataset was retrieved from Kaggle and loaded into the project.  

2. **Handle Missing Values**:  
   - Columns with excessive missing values were dropped.  
   - Relevant missing values were imputed based on statistical techniques.  

3. **Feature Engineering**:  
   - Created **new derived features** such as:  
     - `"days_since_last_purchase"`  
     - `"purchase_frequency"`  
   - Applied **one-hot encoding** to categorical variables like `"loyalty_program"`.  

4. **Random Subset Selection**:  
   - A **random sample of 50,000** data points was extracted to optimize computational efficiency while maintaining dataset representativeness.  

5. **Data Normalization**:  
   - Applied **MinMaxScaler** to standardize numerical features for improved ANN performance.  



### **3.3 ANN Model Architecture**  

#### **Input Layer:**  
- Accepts preprocessed numerical customer data as input.  

#### **Hidden Layers:**  
- **Dense Layer (64 neurons, ReLU Activation)**  
- **Dense Layer (128 neurons, ReLU Activation)**  
- **Dense Layer (64 neurons, ReLU Activation, Dropout 0.3 to prevent overfitting)**  

#### **Output Layer:**  
- **Dense (1 neuron, Sigmoid Activation)** – Outputs probability of churn.  



### **3.4 Model Training & Optimization**  

- **Optimizer Used:** Adam  
- **Loss Function:** Binary Crossentropy  
- **Training Epochs:** 50 _(Tunable via Streamlit dashboard)_  
- **Batch Size:** 32  
- **Evaluation Metrics:**  
  ✔ Accuracy  
  ✔ Precision  
  ✔ Recall  
  ✔ F1-score

## **6. Challenges & Recommendations**  

### **Challenges Faced**  
- **Data Imbalance:**  
  - The dataset had significantly fewer churn cases compared to non-churn cases, affecting model learning and prediction accuracy.  
- **Overfitting Issues:**  
  - The model initially showed overfitting when trained with too many epochs, requiring careful regularization.  
- **Feature Importance Interpretation:**  
  - Some features had strong correlations, making it challenging to determine which factors most influenced churn.  
- **Processing Large Datasets:**  
  - Handling real-world datasets with millions of records required efficient memory management and batch processing.  

### **Recommendations for Improvement**  
1. **Address Data Imbalance:**  
   - Apply **oversampling (SMOTE)** or **undersampling techniques** to balance the dataset and improve model learning.  
2. **Enhance Model Generalization:**  
   - Introduce **dropout layers and L2 regularization** to mitigate overfitting while improving prediction stability.  
3. **Feature Engineering & Selection:**  
   - Use **SHAP (SHapley Additive Explanations)** or **Permutation Importance** to rank and select the most impactful features.  
4. **Optimize Hyperparameters:**  
   - Implement **Grid Search and Bayesian Optimization** to find the best combination of neurons, layers, and activation functions.  
5. **Scalability & Deployment:**  
   - Deploy the model as an **API using Flask or FastAPI**, integrating it with customer engagement platforms for real-time churn predictions.
